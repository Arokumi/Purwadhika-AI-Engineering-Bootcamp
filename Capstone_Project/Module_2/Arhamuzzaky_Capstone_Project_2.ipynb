### Data Encoding

```python
# Encoding non-ordinal columns with one-hot encoding

non_ordinal_list = ["OnlineSecurity", "OnlineBackup", "InternetService", "DeviceProtection", "TechSupport"]

# Saya menggunakan drop_first = true agar tidak kena isu Multicollinearity
encoded_df = pd.get_dummies(df, columns=non_ordinal_list, drop_first=True)
encoded_df.head()
```

```python
# Encoding ordinal data with scikit ordinalEncoder

from sklearn.preprocessing import OrdinalEncoder

ordinal_list = "Contract" # Saya membiarkan Tenure dan MonthlyCharges sebab mereka sudah dalam bentuk numerical dan tidak perlu ordinalEncoder.

order_list = ["Month-to-month", "One year", "Two year"]

encoder = OrdinalEncoder(categories=[order_list])

encoded_df[ordinal_list] = encoder.fit_transform(encoded_df[[ordinal_list]])

encoded_df.head()
```

### Feature Selection

Untuk melakukan feature selection saya akan menggunakan dua tools statistics untuk menentukan fitur mana saja yang akan kita pilih dan fitur apa yang akan singkirkan; Hypothesis Testing dan Mutual Information. 

**Menggunakan hypothesis testing**, kita akan menguji masing-masing fitur untuk menentuakn **apa probabilitas bahwa fitur ini tidak berhubungan dengan churn?** Nilai probabilitas ini akan tampil dalam bentuk P-Value, dimana p-value yang rendah mengartikan bahwa **sangat tidak mungkin bahwa tidak ada hubungan antara fitur dan churn**. Secara standar, orang-orang statistik menggunakan p-value 0.05 sebagai nilai default/significance value; Maka:
> Jika p-value > 0.05, mungkin-mungkin saja jika fitur dan churn tidak ada korelasi/independent
> Jika p-value < 0.05, fitur dan churn hampir pasti mempunyai korelasi/dependensi antar satu sama lain.
  
  

**Menggunakan tes Mutual Information**, kita dapat mengukur dependence/ketergantungan antar satu variable dengan variable lain. Dengan Mutual Information, kita dapat tahu uncertainty/ketidakyakinan dari variable y bisa dikurangi berapa banyak jika kita tahu sebauh variable X. Intinya, Mutual Information dapat menunjukkan korelasi antar variable juga.
> Hasil test Mutual Information itu relatif, sehingga kita juga dapat membandingkan antar variable untuk tahu variable manakah yang paling berpengaruh dan paling tidak berpengaruh terhadap churn.

#### Hypothesis Test to Check Correlation

```python
# Correlation Measurement using Point-Biserial and Chi2

import scipy.stats as stats

correlation_frame = pd.DataFrame(columns=["Feature", "Statistic Value", "Probability Value", "Method"])


# Point-Biserial Correlation Coefficient -> "tenure" and "Monthly Charges"

r_tenure, p_tenure = stats.pointbiserialr(scaled_df["Churn"].to_numpy(), scaled_df["tenure"].to_numpy())

r_charge, p_charge = stats.pointbiserialr(scaled_df["Churn"].to_numpy(), scaled_df["MonthlyCharges"].to_numpy())


correlation_frame.loc[len(correlation_frame)] = ["Tenure", r_tenure, p_tenure, "Point-Biserial"]
correlation_frame.loc[len(correlation_frame)] = ["Monthly Charges", r_charge, p_charge, "Point-Biserial"]

correlation_frame.head()
```

> Untuk menghitung korelasi antar data continuous ("Tenure" dan "MonthlyCharges") dengan data discrete boolean/dikotomi ("Churn"), saya menggunakan metode point biserial correlation. Tools ini mirip dengan Pearson Correlation, namun dimana Pearson Correlation perlu dua set data yang continous untuk menghitung value korelasi antar variable, point-biserial bisa menghitung korelasi antar data continuous dan data dikotomi. 
> Statistic value Point-biserial mirip dengan nilai r dari pearson correlation cofficient, dimana jika nilai tersebut menjauh dari angka nol itu menunjukkan ada korelasi antar kedua variable.
> Dari hasil test kita, bisa melihat bahwa tenure mempunyai korelasi negatif terhadap churn; maka dengannya turunnya tenure churn naik.
> Sedangkan monthly charges punya positive correlation; maka jika monthly charge seorang naik, probability untuk churn juga naik.


<br>
  
Untuk hypothesis testing data non-continuous, saya menggunakan chi square untuk test probabilitas jika sebuah fitur independent dari churn.

```python
# Non-linear Correlation Measurement using Chi Square

# Get old data table and remove unused columns
temp_df = df
temp_df = temp_df.drop(columns=["tenure", "MonthlyCharges", "Churn"])

for column in temp_df:
    contingency_matrix = pd.crosstab(temp_df[column], df["Churn"])
    chi_value, p_value, _, _ = stats.chi2_contingency(contingency_matrix)

    correlation_frame.loc[len(correlation_frame)] = [column, chi_value, p_value, "X\u00b2"]

correlation_frame = correlation_frame.sort_values(by="Probability Value", ascending=False)
correlation_frame
```

> Dari hasil perhitungan hypothesis testing, dapat dilihat bahwa nilai rata-rata probability value kita jauh dibawah significance value 0.05. Ini menunjukkan bahwa sangat tidak mungkin jika fitur-fitur kita tidak ada korelasi sama sekali dengan churn. Namun dari semua data fitur kita ada tiga yang punya probability tertinggi, yaitu Dependents, PaperlessBilling, dan MonthlyCharges. Maka diantara semua fitur kita, menurut hypothesis testing tiga itu adalah fitur terlemah.


#### Mutual Information to Check Usefulness/Impact on Results
Nah untuk Mutual Information, tools tersebut butuh data yang tidak dalam bentuk ordinal encoding, sehingga saya membuat temporary dataframe untuk menghitung Mutual Information dimana ada data tenure dan monthlycharges yang telah di scaling + data-data categorical yang di ordinal encoding, bukan one-hot encoding.

```python
# Creating temporary dataframe for Mutual Information test

scaled = scaled_df[["tenure", "MonthlyCharges"]]

encoder = OrdinalEncoder()
temp_df = encoder.fit_transform(temp_df)

temp_df = pd.concat([temp_df, scaled], axis=1)
temp_df.head()
```

```python
# Testing Mutual Information

from sklearn.feature_selection import mutual_info_classif

discrete_mask = [True, True, True, True, True, True, True, True, False, False]


mutual_info = mutual_info_classif(temp_df, encoded_df["Churn"], discrete_features = discrete_mask)
print(pd.Series(mutual_info, index=temp_df.columns))
```

> Diatas adalah hasil dari test Mutual Information. Namun, tes ini tidak berarti apa-apa karena kita tidak punya pembanding untuk mengetahui jika nilai-nilai yang kita dapatkan itu bagus atau nggak. Untuk mengetes ini, kita melakukan tes ulang Mutual Information namun menggunakan target data churn yang telah di permute/campur sehingga random. Sehingga kita punya pembanding antar nilai Mutual Information berkorelasi dengan nilai Mutual Information yang tidak ada korelasi/independen.

```python
# Testing Mutual Information with Permuted Churn

permuted_churn = pd.DataFrame()
permuted_churn["Churn"] = encoded_df["Churn"].sample(frac=1, random_state=42).reset_index(drop=True)

mutual_info = mutual_info_classif(temp_df, permuted_churn.to_numpy().flatten(), discrete_features = discrete_mask)
print(pd.Series(mutual_info, index=temp_df.columns))
```

> Pengujian menggunakan permuted churn menunjukkan nilai yang jauh lebih kecil dibandingkan dengan data normal, yang mengindikasikan bahwa nilai churn tersebut memang memiliki correlation dengan fitur-fitur yang ada. Ketika label churn diacak untuk mensimulasikan kondisi independen, nilainya justru turun signifikan. Dikombinasikan dengan hasil hypothesis testing sebelumnya, kita dapat dengan cukup percaya diri menyimpulkan bahwa terdapat strong correlation antara dataset kita dan data target (churn).

Saya menggunakan cross_validation stratified k fold karena data imbalance antar Churn and Non-churn.
```python
# Creating Functions for Model Usage



# Cross Validation Strategy
cross_validation = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)


def compute_learning_curve(feature_matrix, target_vector, model, model_name, cv, train_sizes=np.linspace(0.1, 1.0, 10), scoring='roc_auc') -> None: 

    train_sizes, train_scores, test_scores = learning_curve(
        model, feature_matrix, target_vector, cv=cv, train_sizes=train_sizes, scoring=scoring, n_jobs=-1
    )

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    # Plotting
    plt.figure(figsize=(7,5))

    # Plotting training scores
    plt.plot(train_sizes, train_scores_mean, marker='o', label='Training ' + scoring)
    plt.fill_between(train_sizes, train_scores_mean-2*train_scores_std, train_scores_mean+2*train_scores_std, alpha=0.15)

    # Plotting cross-validation scores
    plt.plot(train_sizes, test_scores_mean, marker='o', label='CV ' + scoring)
    plt.fill_between(train_sizes, test_scores_mean-2*test_scores_std, test_scores_mean+2*test_scores_std, alpha=0.15)

    # Finalizing plot
    plt.xlabel('Training samples')
    plt.ylabel(scoring)
    plt.title(model_name + " Learning Curve")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
# Creating Evaluation Table for all models

models = {
    "RandomForest (full)": random_forest_model,
    "RandomForest (cut)": random_forest_model_cut,
    "KNN (full)": knn_model,
    "KNN (cut)": knn_model_cut,
    "LogReg (full)": logreg_model,
    "LogReg (cut)": logreg_model_cut,
}

test_map = {
    "full": features_test,
    "cut" : features_test_cut,
}

def which_test(name: str):
    return test_map["cut"] if "(cut)" in name else test_map["full"]

def evaluate_model(name, model, test, target):
    y_pred = model.predict(test)

    try:
        y_proba = model.predict_proba(test)[:, 1] # Probability to churn
        roc_auc = roc_auc_score(target, y_proba)
    except Exception:
        roc_auc = np.nan

    report = classification_report(target, y_pred, output_dict=True, zero_division=0)

    _confusion_matrix = confusion_matrix(target, y_pred)
    tn, fp, fn, tp = _confusion_matrix.ravel()

    row = {
        "Model": name,
        "Accuracy": report["accuracy"],
        "ROC_AUC": roc_auc,

        # Positive class (Churn)
        "Prec(1)": report["1.0"]["precision"],
        "Rec(1)": report["1.0"]["recall"],
        "F1(1)": report["1.0"]["f1-score"],

        # Negative class (Not Churn)
        "Prec(0)": report["0.0"]["precision"],
        "Rec(0)": report["0.0"]["recall"],
        "F1(0)": report["0.0"]["f1-score"],

        # Aggregates / Averages
        "Prec(w)": report["weighted avg"]["precision"],
        "Rec(w)": report["weighted avg"]["recall"],
        "F1(w)": report["weighted avg"]["f1-score"],
        "Support": int(report["macro avg"]["support"]),

        # Confusion matrix values
        "TP": int(tp),
        "TN": int(tn),
        "FP": int(fp),
        "FN": int(fn),
    }
    return row

rows = []
for name, model in models.items():
    dataset_test = which_test(name)
    rows.append(evaluate_model(name, model, dataset_test, target_test))

summary = pd.DataFrame(rows).sort_values(by=["ROC_AUC","F1(w)"], ascending=False).reset_index(drop=True)
summary
```
